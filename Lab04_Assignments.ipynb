```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report

#define kNN classifire analysis method or function with paramaters data, test_size, n_neighbors.
def knn_classifier_analysis(data, test_size, n_neighbors, target_variable):
    

    # Separate features and target variable (taking LABEL as Target variable)
    X = data.drop(columns=target_variable)
    y = data[target_variable]

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Initialize a K-Nearest Neighbors classifier
    knn_classifier = KNeighborsClassifier(n_neighbors=n_neighbors)

    # Train the model
    knn_classifier.fit(X_train, y_train)

    # Predict on training and test data
    y_train_pred = knn_classifier.predict(X_train)
    y_test_pred = knn_classifier.predict(X_test)

    # Find confusion matrix for training and test data
    conf_matrix_train = confusion_matrix(y_train, y_train_pred)
    conf_matrix_test = confusion_matrix(y_test, y_test_pred)

    # Print confusion matrices
    print("Confusion Matrix - Training Data:")
    print(conf_matrix_train)

    print("\nConfusion Matrix - Test Data:")
    print(conf_matrix_test)

    # Generating classification report for training and test data
    print("\nClassification Report - Training Data:")
    print(classification_report(y_train, y_train_pred))

    print("\nClassification Report - Test Data:")
    print(classification_report(y_test, y_test_pred))

#initiling the test size, k-neighbours and target variable
test_size = 0.2
n_neighbors = 3
target_variable = 'LABEL'

#Calling function
knn_classifier_analysis(data, test_size, n_neighbors, target_variable)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
import numpy as np

def linear_regression_analysis(data, target_variable, test_size):
    
    # Separate features and target variable
    X_reg = data.drop(columns=target_variable)
    y_reg = data[target_variable]

    # Split the data into training and test sets
    X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=test_size, random_state=42)

    # Initialize a Linear Regression model
    regressor = LinearRegression()

    # Train the model
    regressor.fit(X_train_reg, y_train_reg)

    # Predict on test data
    y_pred_reg = regressor.predict(X_test_reg)

    # Calculate metrics
    mse = mean_squared_error(y_test_reg, y_pred_reg)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_test_reg, y_pred_reg)
    r2 = r2_score(y_test_reg, y_pred_reg)

    # Print out metrics
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Root Mean Squared Error (RMSE): {rmse}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape}")
    print(f"R^2 Score: {r2}")

#initiling the test_size and target variable
target_variable = 'LABEL'
test_size = 0.2
linear_regression_analysis(data, target_variable, test_size)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def scatter_of_random_data(n_points=20):
    # Generate random data points for features X and Y
    np.random.seed(42)
    X_random = np.random.uniform(1, 10, n_points)
    Y_random = np.random.uniform(1, 10, n_points)

    # Assign classes based on a simple rule: Class 0 if X < 5, Class 1 otherwise
    classes = np.where(X_random < 5, 0, 1)

    # Scatter plot of the training data
    plt.figure(figsize=(8, 6))
    plt.scatter(X_random[classes == 0], Y_random[classes == 0], color='blue', label='Class 0')
    plt.scatter(X_random[classes == 1], Y_random[classes == 1], color='red', label='Class 1')
    plt.xlabel('Feature X')
    plt.ylabel('Feature Y')
    plt.title('Scatter Plot of Training Data')
    plt.legend()
    plt.show()

#Function calling
scatter_of_random_data(n_points=20)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

def knn_classification_and_plot(n_neighbors=3, n_points=20, grid_range=(0, 10), step=0.1):
    # Generate random training data for features X and Y
    np.random.seed(42)
    X_random = np.random.uniform(1, 10, n_points)
    Y_random = np.random.uniform(1, 10, n_points)

    # Assign classes based on a simple rule: Class 0 if X < 5, Class 1 otherwise
    classes = np.where(X_random < 5, 0, 1)

    # Train kNN classifier with training data
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(np.vstack((X_random, Y_random)).T, classes)

    # Generate test set data
    X_test_knn = np.arange(grid_range[0], grid_range[1], step)
    Y_test_knn = np.arange(grid_range[0], grid_range[1], step)
    X_test_grid, Y_test_grid = np.meshgrid(X_test_knn, Y_test_knn)
    X_test_flat = X_test_grid.flatten()
    Y_test_flat = Y_test_grid.flatten()

    # Combine X and Y into a single test set
    test_set = np.vstack((X_test_flat, Y_test_flat)).T

    # Classify test set data
    test_pred = knn.predict(test_set)

    # Scatter plot of the test data classified by kNN
    plt.figure(figsize=(8, 6))
    plt.scatter(X_test_flat[test_pred == 0], Y_test_flat[test_pred == 0], color='blue', label='Class 0')
    plt.scatter(X_test_flat[test_pred == 1], Y_test_flat[test_pred == 1], color='red', label='Class 1')
    plt.xlabel('Feature X')
    plt.ylabel('Feature Y')
    plt.title('Scatter Plot of Test Data (kNN Classification)')
    plt.legend()
    plt.show()

#Function calling
knn_classification_and_plot(n_neighbors=3, n_points=20, grid_range=(0, 10), step=0.1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

def knn_classification_multiple_k(ks, n_points=20, grid_range=(0, 10), step=0.1):
    # Generate random training data for features X and Y
    np.random.seed(42)
    X_random = np.random.uniform(1, 10, n_points)
    Y_random = np.random.uniform(1, 10, n_points)

    # Assign classes based on a simple rule: Class 0 if X < 5, Class 1 otherwise
    classes = np.where(X_random < 5, 0, 1)

    # Generate test set data
    X_test_knn = np.arange(grid_range[0], grid_range[1], step)
    Y_test_knn = np.arange(grid_range[0], grid_range[1], step)
    X_test_grid, Y_test_grid = np.meshgrid(X_test_knn, Y_test_knn)
    X_test_flat = X_test_grid.flatten()
    Y_test_flat = Y_test_grid.flatten()

    # Combine X and Y into a single test set
    test_set = np.vstack((X_test_flat, Y_test_flat)).T

    # Set up the plot for multiple subplots
    plt.figure(figsize=(12, 10))

    # Repeat classification for different values of k
    for i, k in enumerate(ks):
        # Train kNN classifier with the specified k
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(np.vstack((X_random, Y_random)).T, classes)

        # Classify test set data
        test_pred = knn.predict(test_set)

        # Create subplot for each k
        plt.subplot(2, 2, i + 1)
        plt.scatter(X_test_flat[test_pred == 0], Y_test_flat[test_pred == 0], color='blue', label='Class 0')
        plt.scatter(X_test_flat[test_pred == 1], Y_test_flat[test_pred == 1], color='red', label='Class 1')
        plt.xlabel('Feature X')
        plt.ylabel('Feature Y')
        plt.title(f'K = {k}')
        plt.legend()

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()

# Example usage:
ks = [1, 3, 5, 7]
knn_classification_multiple_k(ks, n_points=20, grid_range=(0, 10), step=0.1)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

def knn_grid_search(file_path, target_variable, param_range=np.arange(1, 11), test_size=0.2):
    # Load the dataset
    data = pd.read_csv(file_path)

    # Separate features and target variable
    X = data.drop(columns=target_variable)
    y = data[target_variable]

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Set up the parameter grid for 'k'
    param_grid = {'n_neighbors': param_range}

    # Perform grid search with cross-validation
    grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    # Output the best 'k' and the corresponding score
    best_k = grid_search.best_params_['n_neighbors']
    best_cv_score = grid_search.best_score_
    print(f"Best k: {best_k}")
    print(f"Best Cross-Validation Score: {best_cv_score}")

    # Train the kNN classifier with the best 'k' found by grid search
    best_knn = KNeighborsClassifier(n_neighbors=best_k)
    best_knn.fit(X_train, y_train)

    # Predict on the test set
    y_test_best_pred = best_knn.predict(X_test)

    # Calculate and print test accuracy
    test_accuracy = accuracy_score(y_test, y_test_best_pred)
    print(f"Test Accuracy with best k ({best_k}): {test_accuracy}")

# Example usage:
file_path = 'DCT_withoutduplicate 5.csv'
target_variable = 'LABEL'  # Replace with the actual target variable name
knn_grid_search(file_path, target_variable)
```
